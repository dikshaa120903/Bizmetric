```python
#list Comphrension
#1.Write a list comprehension to generate squares of numbers from 1 to 10
squares = [x**2 for x in range(1, 11)]
print(squares)
```

```python
#2. Create a list of even numbers between 1 and 50 using list comprehension
even_numbers = [x for x in range(1, 51) if x % 2 == 0]
print(even_numbers)
```

```python
#3. Convert all strings in a list to uppercase using list comprehension.
my_list = ["diksha", "anuj", "girir", "anjali"]

upper_list = [item.upper() for item in my_list]

print("Original List:", my_list)
print("Uppercase List:", upper_list)
      
```

```python
#4. Given a list of integers, create a new list that contains only the positive numbers.

numbers = [-10, 5, -3, 8, 0, -2, 15]
positive_numbers = [num for num in numbers if num > 0]

print("Original List:", numbers)
print("Positive Numbers:", positive_numbers)
```

```python
#5. Create a list of tuples (num, num^2) for numbers 1 to 5.
result = [(num, num**2) for num in range(1, 6)]

print(result)
```

```python
#6. Extract all vowels from a given string using list comprehension.
text = "Diksha Malusare"

vowels = [ch for ch in text if ch.lower() in "aeiou"]

print(vowels)
```

```python
#7. Flatten a 2D list using list comprehension.
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [item for row in matrix for item in row]
print("Flattened List:", flattened)
```

```python
#8. Replace all negative numbers in a list with 0 using list comprehension. 
numbers = [-5, 3, -2, 7, -1]

updated = [num if num >= 0 else 0 for num in numbers]

print(updated)
   
```

```python
#9. Given a list of words, create a list of lengths of each word.
words = ["giri", "anuj", "aujali", "diksha"]

lengths = [len(word) for word in words]

print(lengths)
```

```python
#10. Filter out words that start with the letter 'A' or 'a'. 
words = ["Apple", "banana", "Avocado", "grapes", "apricot"]

filtered = [word for word in words if word.lower().startswith("a")]

print(filtered)
```

```python
#11. From a list of numbers, generate a list of “even” or “odd” strings using list comprehension. (Like → [“even”, “odd”, “odd”, “even”…]
numbers = [1, 2, 3, 4, 5, 6]

result = ["even" if num % 2 == 0 else "odd" for num in numbers]

print(result)
```

```python
#15. Extract numeric digits from a string using list comprehension.
text = "Diksha123Malusare456"   
digits = [ch for ch in text if ch.isdigit()]
print(digits)
```

```python
#16. Use list comprehension to remove all spaces from a string.
text = "Diksha Malusare"
no_spaces = ''.join([ch for ch in text if ch != ' '])   
print(no_spaces)
```

```python
#17. Create a list of characters that appear more than once in a string.
text = "Diksha Malusare"
char_count = {}
for char in text:
    char_count[char] = char_count.get(char, 0) + 1
repeated_chars = [char for char, count in char_count.items() if count > 1]
print(repeated_chars)
```

```python
#18. From a list of sentences, generate a list of all words (split using list comprehension).
sentences = ["Hello world", "This is a test", "Python is great"]
all_words = [word for sentence in sentences for word in sentence.split()]
print(all_words)
```

```python
#9. Create a list of unique elements from a list using list comprehension + condition.
original_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = [x for i, x in enumerate(original_list) if x not in original_list[:i]]
print(unique_list)
```

```python
#20. Generate all pairs (x, y) where x is from list A and y is from list B (cartesian product).
lA = [1, 2, 3]
lB = ['a', 'b', 'c']
pairs = [(x, y) for x in lA for y in lB]
print(pairs)
```

```python
#12. Use lambda to compute factorial using reduce() (yeah, that one-liner madness).
from functools import reduce
factorial = lambda n: reduce(lambda x, y: x * y, range(1, n + 1)) if n > 0 else 1
print(factorial(5)) 
```

```python
#20. Write a lambda to remove duplicates from a list using filter + set.
numbers = [1, 2, 2, 3, 4, 4, 5]
seen = set()
unique_numbers = list(filter(lambda x: not (x in seen or seen.add(x)), numbers))
print(unique_numbers)
```
