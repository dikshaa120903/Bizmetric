```python
#2. Create a list of even numbers between 1 and 50 using list comprehension
en = [x for x in range(1, 51) if x % 2 == 0]
print(en)
```

```python
#3. Convert all strings in a list to uppercase using list comprehension.
l = ["diksha", "anuj", "girir", "anjali"]

ul = [item.upper() for item in l]

print("original list:", l)
print("uppercase list:", ul)
      
```

```python
#4. Given a list of integers, create a new list that contains only the positive numbers.

n = [-10, 5, -3, 8, 0, -2, 15]
pn = [num for num in n if num > 0]

print("Original List:", n)
print("Positive Numbers:", pn)
```

```python
#17. Create a list of characters that appear more than once in a string.
t = "Diksha Malusare"
char_count = {}
for char in t:
    char_count[char] = char_count.get(char, 0) + 1
repeated = [char for char, count in char_count.items() if count > 1]
print(repeated)
```

```python
#9. Create a list of unique elements from a list using list comprehension + condition.
original_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = [x for i, x in enumerate(original_list) if x not in original_list[:i]]
print(unique_list)
```
