```python
#14
books = {
    "1st-4th": {"Hindi": 60, "Marathi": 60, "English": 80, "Science": 90, "Maths": 100},
    "5th-8th": {"Hindi": 100, "Marathi": 100, "English": 100, "Science": 120, "Maths": 140},
    "9th-10th": {"Hindi": 150, "Marathi": 150, "English": 150, "Science": 200, "Maths": 250}
}

notebooks = {
    "square": {100: 40, 200: 70},
    "4_lines": {100: 30, 200: 50},
    "2_lines": {100: 30, 200: 50},
    "single_lines": {100: 60, 200: 100},
    "A4": {100: 100, 200: 180}
}

total = 0
bill = []

while True:
    std = int(input("\nEnter class (1–10): "))

    if 1 <= std <= 4:
        standard = "1st-4th"
    elif 5 <= std <= 8:
        standard = "5th-8th"
    elif 9 <= std <= 10:
        standard = "9th-10th"
    else:
        print("Invalid class")
        continue

    subject = input("Enter subject: ")
    if subject not in books[standard]:
        print("Invalid subject")
        continue

    notebook_type = input("Enter notebook type: ")
    if notebook_type not in notebooks:
        print("Invalid notebook type")
        continue

    pages = int(input("Enter pages (100 or 200): "))
    if pages not in notebooks[notebook_type]:
        print("Invalid page count")
        continue

    book_price = books[standard][subject]
    notebook_price = notebooks[notebook_type][pages]
    amount = book_price + notebook_price

    total += amount
    bill.append((standard, subject, notebook_type, pages, amount))

    choice = input("Do you want to add more items? (yes/no): ")
    if choice.lower() != 'y':
        break

# ---------- FINAL BILL ----------
print("\n=========== FINAL BILL ===========")
for item in bill:
    print(f"Standard: {item[0]}, Subject: {item[1]}, "
          f"Notebook: {item[2]} ({item[3]} pages) → {item[4]}")

print("---------------------------------")
print("TOTAL AMOUNT =", total)
```

```python
#35. List2 = [2,70,'work', para, 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3,10,302.5]
def multiply_by_two(item):

    if isinstance(item, (int, float)):
        return item * 2

    elif isinstance(item, str):
        return item * 2

    elif isinstance(item, list):
        return [i * 2 for i in item]

    elif isinstance(item, tuple):
        return tuple(i * 2 for i in item)

    elif isinstance(item, set):
        return {i * 2 for i in item}

    elif isinstance(item, dict):
        return {k: v * 2 for k, v in item.items()}

    else:
        return item


List2 = [21, 701, 'diksha', 'anuj', 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'},302.5]

result = [multiply_by_two(x) for x in List2]

print("After Multiplying by 2:", result)

```

```python
#36. List2 = [2,70,'work', para, 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3,10,302.5]
#Multiply each and every element from list2 by 2 and store the answer in list3
list2 = [21, 701, 'diksha', 'anuj', 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 302.5]

list3 = []

for item in list2:

    # Numbers
    if isinstance(item, (int, float)):
        list3.append(item * 2)

    # Strings
    elif isinstance(item, str):
        list3.append(item * 2)

    # List
    elif isinstance(item, list):
        list3.append([i * 2 for i in item])

    # Tuple
    elif isinstance(item, tuple):
        list3.append(tuple(i * 2 for i in item))

    # Set
    elif isinstance(item, set):
        list3.append({i * 2 for i in item})

    # Dictionary
    elif isinstance(item, dict):
        new_dict = {}
        for k, v in item.items():
            new_dict[k] = v * 2
        list3.append(new_dict)

print("List2:", list2)
print("List3:", list3)
```
